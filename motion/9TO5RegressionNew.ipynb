```python
def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))
def train_predict(train_X,train_y,test_X,test_y):
    print('============================predicting==============================')
    stack = StackingCVRegressor(regressors=(lasso,rf,xgb,ENet, KRR, lgb, gbm), 
                              meta_regressor=ENet)

    stack.fit(train_X, train_y)
    #sclf_pred = sclf.predict(X_train.values)
    stack_train_pred = stack.predict(train_X)
    stack_pred = stack.predict(test_X)
    print('mean_squared_error:')
    print(rmsle(train_y, stack_train_pred))
    return stack_pred

output=[]
t=1
for i in ['Hand_L','Hand_R','Elbow_L','Elbow_R','Forearm_L','Forearm_R','Shoulder_L','Shoulder_R','Shoulder_pitch_L','Shoulder_pitch_R']:
#for i in ['Shoulder_L','Shoulder_R','Shoulder_pitch_L','Shoulder_pitch_R']: #'Hand_L','Hand_R' 'Elbow_R','Forearm_R','Shoulder_R','Shoulder_pitch_R'
    location = i
    print('the %d th location' % (t))
    t+=1
    train_X,train_y,test_X,test_y = load_data(location)
    stack_pred = train_predict(train_X,train_y,test_X,test_y)
    plt.figure()
    plt.plot(stack_pred, label = 'predict')
    plt.plot(test_y, label = 'true')
    plt.legend()
    plt.show()
    #inv_pred,inv_y = inverse_data(stack_pred,val,val_y,scaler)
    output.append(stack_pred*5600)
```
